```python
#闭包
def func_closure():
    def geat(message):

        print(f'This is {message}')
    return geat

send_message = func_closure()
send_message('he')
```

```python
#装饰器
import functools

def func_closure(func):
    @functools.wraps(func)
    def geat(*args, **kwargs):
        print('This is func')
        return func(*args, **kwargs)
    return geat

@func_closure
def message(message):
    print(f'This is {message}')
    
message('he')
```
